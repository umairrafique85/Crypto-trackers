```r
library(httr)
library(jsonlite)
library(anytime)

get_markets <- function(fcurr, tocurr, minvol) {
    
    
    url <- "https://api.cryptonator.com"
    path <- sprintf("api/full/%s-%s", fcurr, tocurr)
    raw.result <- GET(url=url, path=path)
    raw.content <- rawToChar(raw.result$content)
    parsed.content <- fromJSON(raw.content)
    
    diffmarkets <- parsed.content$ticker$markets
    diffmarkets$price <- round(as.numeric(diffmarkets$price))
    diffmarkets$volume <- round(diffmarkets$volume)
    
    with(diffmarkets, plot(volume~price, pch=19, col='blue'))
    with(diffmarkets, text(volume~price, labels = market, pos = 1, cex = 0.7))
    
    diffmarkets <- diffmarkets[diffmarkets$volume>=minvol,]
    
    maxprice <- diffmarkets[diffmarkets$price==max(diffmarkets$price),]
    minprice <- diffmarkets[diffmarkets$price==min(diffmarkets$price),]
    
    
    sprintf("The maximum price recorded at %s was %d in %s, which had a trading volume of %d. The minimum was %d in %s, with a trading volume of %d", 
            anytime(parsed.content$timestamp), maxprice$price, maxprice$market, maxprice$volume, 
            minprice$price, minprice$market, minprice$volume)
    
}

# Usage example: get_markets('btc', 'usd', 80)
# First two arguments are self explanatory. The last one is minimum trade volume. To filter out results from exchanges with 
# less trade volume than specified.
```

```r
get_markets('btc', 'usd', 80)
```
